```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("Accident.csv")
print("Accident Dataset:")
print(df)
plt.figure(figsize=(3,2))
df['City'].value_counts().plot(kind='bar',color='skyblue')
plt.title('Accident by city')
plt.xlabel('city')
plt.ylabel('Number of Accidents ')
plt.tight_layout()
plt.show()
plt.figure(figsize=(3,2))
df['Weather'].value_counts().plot(kind='pie',autopct='%1.1f%%',colors=['lightcoral','gold','lightblue','lightgreen'])
plt.title('Weather Condition')
plt.ylabel('')
plt.tight_layout()
plt.show()
plt.figure(figsize=(3,2))
df['Time of Day'].value_counts().plot(kind='bar',color='orchid')
plt.title('Accidents by Time of Day')
plt.xlabel('Time of Day')
plt.ylabel('Number of Accidents')
plt.tight_layout()
plt.show()
print("\n Summary Insights:")
print("Most accidents happened in:",df['City'].mode()[0])
print("Most common weather during accidents:",df['Weather'].mode()[0])
print("Most accident-prone time of day:",df['Time of Day'].mode()[0])
```
